```python
# Comparision between List and Array

L=[1,2,3]
A=np.array([1,2,3])

for x in L:
    print(x)
for y in A:
    print(y)
    
# Appending and Concatenating is possible for 'Lists' but it cannot use for 'Arrays'

L.append(4)
# result: [1, 2, 3, 4]
A.append(4) 
# result: This leads to an Error

L = L+[5]
# result: [1, 2, 3, 4, 5]
A= A+ [5]
# result: array([6, 7, 8])

# Multiplying also has a different result in each 

print(2* L) 
# result: [1, 2, 3, 1, 2, 3]
print(2* A)
# result: array([2, 4, 6])

# Conclusion: 
# List works like a generate data structure, however Array do the math on the elements.
```

```python
# Dote Products

a= np.array([1,2])
b= np.array([3,4])
dot_1=0
dot_2=0

# 1
for x,y in zip(a,b):
    dot_1 += x*y
print(dot_1)

# 2
for i in range(len(a)):
    dot_2 += a[i] * b[i]
print(dot_2)

# 3
np.sum(a*b)

# 4 
(a*b).sum()

# 5 
np.dot(a,b)

# 6
a.dot(b)

# 7
a @ b
```

```python
# Matrices 

L=[[1,2],[3,4]]
print(L[0][1])

M1 = np.array([[1,2],[3,4]])
print(M1[0,1]) # or M1[0][1]

print(M1.T)
print(np.diag(M1))
print(np.diag([1,4]))
print(np.linalg.inv(M1))
print(np.trace(M1))
print(np.linalg.eig(M1))
```

```python
# Exercise 2: MatPlot 

x1=np.random.uniform(-1,1,1000)
x2=np.random.uniform(-1,1,1000)

for i in range(len(x1)):
    if x1[i] * x2[i] <= 0:
        plt.scatter(x1[i],x2[i], c='red')
    else:
        plt.scatter(x1[i],x2[i], c='blue')

plt.axis([-1,1,-1,1])
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('XOR Plot')
```

```python
# Scipy: PDF and CDF distribution
# PDF

x= np.linspace(-6,6,1000)
fx= norm.pdf(x, loc=0, scale=1)
plt.plot(x,fx);
```

```python
# CDF

fy= norm.cdf(x, loc=0, scale=1)
plt.plot(x,fy);
```

```python
# LogPDF

flog= norm.logpdf(x, loc=0, scale=1)
plt.plot(x,flog);
```

```python
# Convolution

from PIL import Image
from scipy.signal import convolve2d
im = Image.open('Documents/مدارک/ed3.png')
arr = np.array(im)
gray = np.mean(im, axis=2)
y= np.linspace(-6,6,50)
FX= norm.pdf(y, loc=0, scale=1)
filt=np.outer(FX,FX)
out= convolve2d(gray,filt)

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(out, cmap='gray')
```

```python
# Exercise 4: Scipy

from PIL import Image
from scipy.signal import convolve2d

im = Image.open('lena_gray.jpg')
arr = np.array(im)
gray = np.mean(im, axis=2)

hx= np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
hy= np.array([[1,2,1],[0,0,0],[-1,-2,-1]])

gx= convolve2d(gray,hx)
gy= convolve2d(gray,hy)

G= np.sqrt((gx**2)+(gy**2))
plt.imshow(G, cmap='gray');
```
